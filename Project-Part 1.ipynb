```octave
fin = fopen('data8OX.txt');
d = 8;
n = 45;
fgetl(fin);fgetl(fin);fgetl(fin);
```

```octave
A = fscanf(fin, '%f', [d+1, n]);
```

```octave
A = A';
```

```octave
X = A(:,1:d);
```

### a. Mean and covariance

```octave
u = sum(X)./n
```

```octave
C = totalCovariance(X,d,u,n)
```

### b. Eigenvalues of C

```octave
E = eig(C)
```

### c. Caclulating percentages

```octave
p = (zeros(1,d));
```

```octave
% compute percentages
for x=1:d
    p(x) = sum(E(1:x)) / sum(E(1:d));
end
```

```octave
p
```
